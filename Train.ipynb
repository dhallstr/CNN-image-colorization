```python
weights_file="flower_colorizer_rmse.pt"

model = Colorizer()

if path.exists(weights_file):
    
    #load model from file
    model.load(weights_file)
    
else:
    model.train_from_dir("imgs/flowers", learning_rate=0.001, num_epochs=200, batch_size=100, model_filename=weights_file)
```
